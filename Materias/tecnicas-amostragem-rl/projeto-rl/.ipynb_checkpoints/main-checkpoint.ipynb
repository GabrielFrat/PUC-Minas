```python
import pandas as pd
import numpy as np
import zipfile
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Carregando dados")
df = pd.read_csv(
    'reclamacoes_anatel_limpa.csv', 
    sep=',', 
    engine='pyarrow',     
    dtype_backend='pyarrow'         
)

df.head()
```

```python
# Tratamento da base de dados

df = df.drop(columns=[''], errors='ignore')
df.head()
```

```python
print(df.describe(include='all'))
```

```python
from scipy.stats import chi2_contingency

df_sample = df.sample(200_000, random_state=42)


def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))

cat_cols = ['UF', 'Marca', 'Serviço', 'CanalEntrada', 'Condição', 'TipoAtendimento']
cramers_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols)

for c1 in cat_cols:
    for c2 in cat_cols:
        cramers_matrix.loc[c1, c2] = cramers_v(df[c1], df[c2])

plt.figure(figsize=(10,7))
sns.heatmap(cramers_matrix.astype(float), annot=True, cmap="Blues")
plt.show()
```

```python
cols = ['UF','Marca','Serviço','Assunto','CanalEntrada','Condição','TipoAtendimento']

for c in cols:
    df[c] = df[c].astype(str)
```

```python
import pandas as pd
import statsmodels.api as sm

cols_agrupamento = ['Ano', 'Mês', 'UF', 'Marca', 'Serviço']

df_agrupado = df.groupby(cols_agrupamento)['SOLICITAÇÕES'].sum().reset_index()
print(f"Novas linhas para análise: {df_agrupado.shape[0]}")
print(df_agrupado.head())

y = df_agrupado['SOLICITAÇÕES']
X = df_agrupado.drop(columns=['SOLICITAÇÕES'])

X_tratado = pd.get_dummies(X, columns=['UF', 'Marca', 'Serviço'], drop_first=True, dtype=int)
X_tratado = X_tratado.astype(float)
X_tratado = sm.add_constant(X_tratado)
modelo = sm.OLS(y, X_tratado).fit()

print(modelo.summary())
```
